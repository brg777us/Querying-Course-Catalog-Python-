```python
#Creating the dictionary from the text file accouting.txt
aclass = {}
acct = open('accounting.txt','r')
classno = acct.readline().strip()
while classno!='':        
        classna = acct.readline().rstrip()
        classcr = acct.readline().rstrip()
        aclass[classno] = classna, classcr
        classno = acct.readline().strip()
acct.close()

#Creating the function that prints the entire dictionary
def coursespec():
    course = input("\nPlease enter course number ")
    try:
        print(course,*aclass[course],sep="\t")
    except KeyError:
        print("\nThe course number entered does not exist.")

#main program
stopmain = "no"
stop2 = "no"
stop3 = "no"
while stopmain == "no":
    question = input("Do you want to see the entire catalog? (Y/N) \n")
    if question == "Y" and stopmain == "no":
        for r in aclass:
            print(r,*aclass[r],sep="\t")
        while stop2 == "no":
            question2 = input("\nDo you want to check a specific course? (Y/N) ")
            if question2 == "N":
                stop2 = "yes"
                stopmain = "yes"
            elif question2 == "Y":
                coursespec()
            else:
                print("Please type either 'Y' or 'N' only please")
    elif question == "N" and stopmain == "no":
        while stop3 == "no":
            question3 = input("\nDo you want to check a specific course? (Y/N) ")
            if question3 == "Y":
                coursespec()
            elif question3 == "N":
                stop3 = "yes"
                stopmain = "yes"
            else:
                print("Please type either 'Y' or 'N' only please")
    else:
        print("Please type either 'Y' or 'N' only please")
```
